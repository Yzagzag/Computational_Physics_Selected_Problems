```python
import random 
import numpy as np 
import math
import matplotlib.pyplot as plt 
import scipy as sp
from scipy import interpolate
```

A script to perform Lagrange interpolation on cross sections measured for the resonant scattering of neutrons from a nucleus with one polynomial. We fit all nine of the data points we obtain with an 8th-order polynomial. We use scipy.interpolate.lagrange. We plot the points and the polynomial from the interpolation.

```python
E_i=[0,25,50,75,100,125,150,175,200]
gE_i=[10.6,16.0,45.0,83.5,52.8,19.9,10.8,8.25,4.7]

test=np.arange(2000)*0.1

gE=interpolate.lagrange(E_i,gE_i)

gE_poly=gE(test)

plt.plot(test, gE_poly,label='Poly Interpolate')
plt.plot(E_i,gE_i,linestyle='',marker='.',ms=10,label='Interp Points')
plt.title('Neutron Scattering Analysis with Interpolation')
plt.xlabel('E_i(MeV)')
plt.ylabel('g(E_i) (mb)')
plt.legend(loc='lower right',frameon=False)
```

We find the resonance energy E_r (the peak position): E_r ~ 75 Mev. This is close to 78 MeV predicted by taking the derivative of gE_poly.
We also find Γ (the full-width at half-maximum): ~60 MeV. This is close to 55 MeV.
