```python
import random 
import numpy as np 
import matplotlib.pyplot as plt
```

This is a program to simulate radioactive decay

Here is a plot of the log of the number of particles left $lnN(t)$ and the log of the decay rate $ln(\nabla N(t)/\nabla t)$ vs. time.

```python
#We plot the decay rate and the number of particles left 
#We input the decay rate, the time alloted, and the Max number of particles
def exponentialDecay(lambda1,maxN,time_max):
    
    seed =68111
    prior = number = nloop = maxN

    for time in np.arange (0 , time_max + 1) :
        for atom in np.arange(1, number + 1 ):
            decay = random.random() 
            if (decay < lambda1): 
                nloop = nloop - 1
        number = nloop
        plt.title('Spontaneous Decay')
        plt.xlabel('Time')
        plt.ylabel('Number Atoms Left')
        if number == 0:
            x=0
        else:
            x=np.log(number)
            
        plt.plot(time, x,'bo')
        plt.plot(time, np.log(prior - number), 'ro')
        number = number
    print ('Blue = ln(Remaining Particles)')
    print ('Red = ln(Decay Rate)')  
    plt.show()
```

```python
exponentialDecay(0.005,1000,500)
```

```python
#We plot the rate of change of the remaining atoms to 
#show it is independent of choice for starting N 
def exponentialDecayIndependenceNslopevsNmax(lambda1,maxN,time_max):
    seed =68111
    
    for N in range(maxN):
        prior=number = nloop = N
        for time in np.arange (0 , time_max + 1) :
            for atom in np.arange(1, number + 1 ):
                decay = random.random() 
                if (decay < lambda1): 
                    nloop = nloop-1
            number = nloop
            plt.title('Slope N(t) vs N starting Choice')
            plt.xlabel('N(0)')
            plt.ylabel('Rate of Change of Remaining Atoms')
            plt.plot(N, (prior-number)/1, 'bo')
            prior = number
    plt.show()
```

```python
exponentialDecayIndependenceNslopevsNmax(.0005,200,50)
```

```python
#We plot the rate of change of the remaining atoms to 
#show it is dependent on the choice for the decay rate
def exponentialDecayDependence(lambdamax,maxN,time_max):
    seed = 68111
    prior = number = nloop = maxN

    for i in np.linspace(0, lambdamax):
        for time in np.arange(0 , time_max + 1) :
            for atom in np.arange(1, number + 1 ):
                decay = random.random() 
                if (decay < i): 
                    nloop = nloop-1
            number = nloop
            plt.title('N(t) vs Lambda')
            plt.xlabel('Lambda')
            plt.ylabel('Rate of Change of Remaining Atoms')
            plt.plot(i, (prior-number)/1, 'bo')
            prior = number
    plt.show()
```

```python
exponentialDecayDependence(0.1,200,20)
```
